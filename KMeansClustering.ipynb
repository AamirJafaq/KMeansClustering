```python
plt.scatter(x=data["cgpa"], y=data["iq"], alpha=0.5)
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.title("CGPA vs IQ")
plt.show()
```

```python
print("From the scatter plot, it can be seen that the data has four cluster. However, we will use clustering \n algorithm to reach the same result")
```

# K-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=4)
model.fit(data)
```

```python
predict=model.predict(data)      
```

```python
print("In above code line, predict method predicts our clusters")
```

```python
data["cluster"]=predict
```

```python
print("In above code line, we append the clustering as new column of data set")
```

```python
clusters=data.groupby("cluster")
cluster1=clusters.get_group(0)
cluster2=clusters.get_group(1)
cluster3=clusters.get_group(2)
cluster4=clusters.get_group(3)
```

```python
print("In obove code line, we separate the data set into four different set. Each data set represents cluster")
```

```python
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.title("CGPA vs IQ")
plt.scatter(cluster1["cgpa"], cluster1["iq"])
plt.scatter(cluster2["cgpa"], cluster2["iq"])
plt.scatter(cluster3["cgpa"], cluster3["iq"])
plt.scatter(cluster4["cgpa"], cluster4["iq"])
```

```python
print("Notice that the clustering is exactly same as the given original data. This is because the data was already in clusters. \nHowever, this confirm that the algorithm works")
```
